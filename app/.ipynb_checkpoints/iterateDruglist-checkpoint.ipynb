```python
def get_drug_list(df):
    sec_dict={'Primary tumor':[],'Metastatic tumor':[],'Undetermined':[]}
    accordion_dict=dict()
    
    for index,row in df.iterrows():

        if row['Primary_site'] not in accordion_dict:
            accordion_dict[row['Primary_site']]=sec_dict.copy()
        if row['Comments'] == 'NaN':
            accordion_dict[row['Primary_site']][row['Tumour_origin']].append([row['cell_line_name'],row['Comments']])
        else:

            accordion_dict[row['Primary_site']][row['Tumour_origin']].append([row['cell_line_name'],'NaN'])

    return accordion_dict
```
